```python
# Load CSV file into a Spark DataFrame
df = spark.read.format("csv").option("header", "true").load("/Volumes/movie/movie_bronze/dataset-movies/Popular Movies TV shows from Prime Videos Netflix version_3.csv")

# Write DataFrame to a Delta table
df.write.format("delta").option("delta.columnMapping.mode", "name").mode("overwrite").saveAsTable(f"""movie.movie_bronze.movies""")

display(spark.table("movie.movie_bronze.movies"))
```
